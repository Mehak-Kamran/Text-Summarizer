```python
from datasets import load_dataset

# Load the CNN/Daily Mail dataset with 12,000 samples
dataset = load_dataset("cnn_dailymail", "3.0.0", split='train[:12000]')

print(f"Number of samples: {len(dataset)}")
print(dataset[0])  # View a sample article
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

# Load BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

def summarize(text):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Load BERT model and tokenizer for classification (34 categories)
num_classes = 34
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes)

def classify(text):
    inputs = bert_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    outputs = bert_model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()
    return predicted_class
```

```python
class_labels = [
    "Politics", "Business", "Entertainment", "Sports", "Technology", "Science",
    "Health", "Travel", "Education", "Lifestyle", "Fashion", "Food", "Environment",
    "Culture", "World", "Economy", "Finance", "Real Estate", "Energy", "Automotive",
    "Retail", "Pharma", "Medicine", "Law", "Weather", "Transportation", "Space",
    "Military", "Agriculture", "Art", "History", "Gaming", "Social Media"
]

for i in range(10):  # Process first 5 articles as an example
    article = dataset[i]["article"]
    summary = summarize(article)
    predicted_class = classify(article)

    print(f"Article {i + 1}:")
    print("Summary:", summary)
    print("Predicted Category:", class_labels[predicted_class])
    print("\n")
```
