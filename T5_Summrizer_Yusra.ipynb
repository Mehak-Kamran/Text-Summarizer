```python
from datasets import load_dataset

# Load the CNN/Daily Mail dataset (12K samples)
dataset = load_dataset("cnn_dailymail", "3.0.0", split='train[:12000]')

# Check the size of the dataset
print(f"Number of samples: {len(dataset)}")

# View a sample
print(dataset[0])
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the T5 model and tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

def summarize(text):
    input_text = "summarize: " + text
    input_ids = tokenizer.encode(input_text, return_tensors="pt", truncation=True, max_length=512)
    summary_ids = model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Load pre-trained BERT model for classification (34 classes)
num_classes = 34
tokenizer_bert = BertTokenizer.from_pretrained("bert-base-uncased")
model_bert = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes)

# Classification function
def classify(text):
    inputs = tokenizer_bert(text, return_tensors="pt", truncation=True, max_length=512)
    outputs = model_bert(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    return predicted_class
```

```python
class_labels = [
    "Politics", "Business", "Entertainment", "Sports", "Technology", "Science",
    "Health", "Travel", "Education", "Lifestyle", "Fashion", "Food", "Environment",
    "Culture", "World", "Economy", "Finance", "Real Estate", "Energy", "Automotive",
    "Retail", "Pharma", "Medicine", "Law", "Weather", "Transportation", "Space",
    "Military", "Agriculture", "Art", "History", "Gaming", "Social Media"
]

for i in range(10):  # Process first 10 articles as an example
    article = dataset[i]["article"]
    summary = summarize(article)
    category = classify(article)

    print(f"Article {i + 1}:")
    print("Summary:", summary)
    print(f"Category: {class_labels[category]}")  # Adjust with category names
    print("\n")
```
